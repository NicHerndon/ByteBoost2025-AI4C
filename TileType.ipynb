# Get some stats for `background` tile and `tissue` tile

```python
import openslide
import numpy as np
```

## Read the slide

```python
slide_path = 'patient_004_node_4.tif'
wsi = openslide.OpenSlide(slide_path)	# read through object
```

## Get stats for two different tiles

### `Tissue` tile

```python
sample_tissue = wsi.read_region((12597,88778),0,(224,224))
sample_tissue.show()
```

```python
npst = np.asarray(sample_tissue)
```

```python
print('Average (red, green, blue): ({}, {}, {})'.format(
    int(np.average(npst[:][:][0])),
    int(np.average(npst[:][:][1])),
    int(np.average(npst[:][:][2]))))
```

```python
print('Min (red, green, blue): ({}, {}, {})'.format(
    int(np.min(npst[:][:][0])),
    int(np.min(npst[:][:][1])),
    int(np.min(npst[:][:][2]))))
```

```python
print('Max (red, green, blue): ({}, {}, {})'.format(
    int(np.max(npst[:][:][0])),
    int(np.max(npst[:][:][1])),
    int(np.max(npst[:][:][2]))))
```

```python
print('Standard deviation (red, green, blue): ({}, {}, {})'.format(
    int(np.std(npst[:][:][0])),
    int(np.std(npst[:][:][1])),
    int(np.std(npst[:][:][2]))))
```
